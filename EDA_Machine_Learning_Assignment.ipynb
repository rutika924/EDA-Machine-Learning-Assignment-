Question 2:

Explain overfitting and underfitting in ML. How can you detect and prevent
them?

Ans:-

In Machine Learning, the main goal of a model is to learn patterns from training data and make accurate predictions on new or unseen data.
Sometimes, the model may learn too much (overfitting) or too little (underfitting) from the data.
These problems can be understood through the bias–variance tradeoff.



1.   Overfitting:

Overfitting happens when a model learns both the actual patterns and the noise or random fluctuations in the training data.
It performs very well on training data but poorly on test or validation data.

Example:
A student memorizes all the answers from a book but fails when new or slightly different questions appear in the exam.

Causes:

*  Model is too complex (too many parameters or layers).

*  Insufficient or noisy data.

*  Training for too many epochs.

*  No regularization applied.

Detection:

*  High training accuracy but low test accuracy.

*  Large gap between training and validation error.

Prevention:

*  Cross-Validation:

Use techniques like k-fold cross-validation to check how the model performs on unseen data.

*  Regularization Techniques:

Apply L1 (Lasso) or L2 (Ridge) regularization to reduce model complexity.

*  Dropout (in Neural Networks):

Randomly drop neurons during training to prevent over-dependence on specific features.

*  Simplify the Model:

Reduce number of layers or parameters.

*  Early Stopping:

Stop training when validation error starts increasing.


2.   Underfitting:

Underfitting occurs when a model is too simple to capture the underlying patterns of the data.
It performs poorly on both training and testing data.

Example:
A student studies very little and cannot answer even the basic questions.

Causes:

* Model is too simple (e.g., linear model for non-linear data).

* Too few features used.

* Too much regularization applied.

* Insufficient training time.

Detection:

* High training error and high test error.

* Model fails to capture trends in training data.

Prevention:

* Increase Model Complexity:

Use more layers or parameters suitable for the data.

* Reduce Regularization:

Lower regularization strength to allow better learning.

* Train Longer:

Increase epochs or iterations.

* Add More Features:

Include meaningful input variables.




Question 3:

 How would you handle missing values in a dataset? Explain at least three
methods with examples.

Ans:-

In any real-world dataset, missing values are common. They occur when no value is recorded for a feature in some observations.
Handling missing values is critical for Machine Learning, as they can reduce model accuracy, create bias, and affect statistical analysis.

Missing values can be handled using different approaches, depending on the amount of missing data, type of data, and model requirements.



1.   Deletion Method (Removing Missing Data):

This method involves removing rows (observations) or columns (features) that contain missing values. It is simple and works well when the proportion of missing data is small.

Types:

* Listwise Deletion: Remove the entire row if it has any missing value.

* Column Deletion: Remove a column if it has too many missing values.

Example:

Suppose a dataset of 1000 employees has a column “Age,” and only 10 rows have missing values. Using listwise deletion, we can remove those 10 rows without losing much information.

Advantages:

* Very simple and quick to apply.

* Keeps data clean without introducing bias from imputation.

Disadvantages:

* May result in loss of valuable data if missing values are many.

* Can reduce dataset size, which may affect model training.


2.   Mean/Median/Mode Imputation

This method replaces missing values with statistical measures like mean, median, or mode of the non-missing values in the same feature.

Mean Imputation: Replace missing numeric values with the average of the available data.

Median Imputation: Replace missing numeric values with the middle value (less affected by outliers).

Mode Imputation: Replace missing categorical values with the most frequent value.

Example:

* Numeric: “Salary” column has values [30000, 35000, missing, 40000]. The mean = (30000 + 35000 + 40000)/3 = 35000. Replace missing with 35000.

* Categorical: “Department” column has values [HR, IT, IT, missing, HR]. The mode = IT. Replace missing with IT.

Advantages:

* Easy to implement.

* Maintains dataset size.

* Works well when missing data is minimal.

Disadvantages:

* Reduces data variability, which can bias model results.

* Mean is sensitive to outliers; median is better in such cases.

3.   Predictive Modeling / Advanced Imputation

This method predicts missing values using other features in the dataset.
Models like Linear Regression, K-Nearest Neighbors (KNN), Random Forests, or iterative imputation can estimate missing data more accurately than simple mean/median methods.

Example:

* Suppose “Salary” is missing in some rows. Using Linear Regression, we can predict salary using other features like Experience, Age, Education, and Department.

* KNN imputer can replace missing values based on similar rows in the dataset.

Advantages:

* More accurate and data-driven than simple imputation.

* Preserves relationships between features.

Disadvantages:

* Computationally more expensive.

* Requires building and validating predictive models.

* May introduce model dependency for imputation.



Question 4:

What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

Ans:-

Definition:

An imbalanced dataset occurs when the number of examples in different classes of a classification problem is not proportional.

In such datasets, the majority class dominates, and the minority class has very few examples.

Machine Learning models trained on imbalanced data often bias predictions toward the majority class, failing to correctly classify minority instances.

Example:

* Fraud detection: 99% non-fraud transactions (majority), 1% fraud transactions (minority).

* Medical diagnosis: 95% healthy patients vs 5% patients with disease.

* A naive model predicting all as majority would have high accuracy but poor real-world performance.

2. Consequences of Imbalanced Datasets:

* Reduced model performance for minority class predictions.

* Accuracy metric becomes misleading.

* Algorithms may ignore the minority class completely.

* Increased risk of biased decision-making in real-world applications.

3. Techniques to Handle Imbalanced Datasets:

    A.  Random Oversampling & Undersampling

      i. Random Oversampling

 Replicate existing examples from the minority class to balance the class distribution.

Practical Example:

* Dataset: 900 non-fraud, 100 fraud transactions.

* Duplicate fraud transactions until both classes have 900 examples.

Advantages:

* Simple and effective for small datasets.

* Prevents the model from ignoring the minority class.

Disadvantages:

* Can cause overfitting, as the same minority samples are repeated.

* Increases training time.

ii. Random Undersampling

 Remove samples from the majority class to match the minority class size.

Practical Example:

* Reduce non-fraud transactions from 900 to 100, matching the 100 fraud transactions.

Advantages:

* Reduces dataset size, improving training efficiency.

* Simple to implement.

Disadvantages:

* Loss of valuable information from majority class.

* May reduce overall model accuracy.

B. Synthetic Minority Oversampling Technique (SMOTE)



Instead of duplicating minority class samples, SMOTE creates synthetic examples by interpolating between existing minority class instances.

Produces more diverse and representative data points for the minority class.

Practical Example:

* 100 fraud transactions vs 900 non-fraud.

* SMOTE generates synthetic fraud examples based on nearest neighbors until classes are balanced.

* The resulting dataset: 900 fraud + 900 non-fraud transactions.

Advantages:

* Reduces overfitting compared to random oversampling.

* Increases minority class diversity.

Disadvantages:

* Can generate noisy or unrealistic examples.

* Less effective if minority samples are very sparse.

C. Using Class Weights in Models


Many ML algorithms (logistic regression, random forest, SVM) allow assigning higher weights to minority class.

The model penalizes misclassification of minority class more heavily during training.

Practical Example:

* Fraud detection model: assign class_weight='balanced' in scikit-learn.

* Misclassifying a fraud transaction incurs a higher penalty than misclassifying a non-fraud transaction.

Advantages:

* No need to modify the dataset size.

* Works well with large datasets.

Disadvantages:

* Requires careful tuning to avoid overcompensation.

Question 5:

 Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

Ans:-

-Definition:

* Feature scaling is a preprocessing technique in Machine Learning that normalizes or standardizes the range of independent variables (features). It ensures that all features contribute proportionally to the model, preventing features with larger numerical ranges from dominating the learning process.

-Importance of Feature Scaling:

* Feature scaling is crucial because many Machine Learning algorithms are sensitive to the magnitude of features.

* In distance-based algorithms like K-Nearest Neighbors (KNN) and Support Vector Machines (SVM), distances between data points are calculated using features. Features with larger values can dominate these calculations, leading to biased predictions.

* In gradient descent-based algorithms like linear regression, logistic regression, or neural networks, unscaled features can cause uneven gradients, slow convergence, or unstable updates.

* Feature scaling also improves model interpretability and enhances the performance of regularization techniques, where feature magnitudes affect penalty terms.

-Scope of Feature Scaling:

* Feature scaling is used in a wide range of Machine Learning tasks, especially:

* Distance-based algorithms: KNN, SVM, clustering (K-Means), where distance metrics are sensitive to feature magnitude.

* Gradient descent-based algorithms: Linear regression, logistic regression, neural networks, deep learning models.

* Regularization methods: Lasso, Ridge, and Elastic Net, where feature magnitudes affect penalties.

* Neural networks: Min-Max scaling is particularly useful when features are input to activation functions like Sigmoid or ReLU.

Techniques for Feature Scaling:

1. Min-Max Scaling:

* Definition:

-Rescales feature values to a fixed range, usually 0 to 1.

* Practical Use:

-Preserves original distribution and is suitable when specific value ranges are required, e.g., neural networks.

* Advantages:

-Simple, keeps all values in a fixed range, works well for bounded inputs.

* Disadvantages:

-Sensitive to outliers; extreme values can shrink the majority of data into a narrow range.

2. Standardization (Z-score normalization):

* Definition:

-Transforms features to have mean = 0 and standard deviation = 1.

* Practical Use:

-Suitable when data is normally distributed or contains outliers. Works well with algorithms like SVM, logistic regression, and gradient descent.

* Advantages:

-Less affected by outliers, maintains relationships between features, ensures balanced gradients.

* Disadvantages:

-Does not bound values to a fixed range; may be less intuitive in interpretation.

- Impact on Machine Learning Algorithms:

1. Distance-Based Algorithms (KNN, SVM):

* These algorithms calculate similarity or distance between points.

* Without scaling, features with larger ranges dominate the distance computation, biasing the model.

* Scaling ensures all features contribute equally to distances, improving neighbor selection in KNN and hyperplane separation in SVM.

2. Gradient Descent-Based Algorithms (Linear Regression, Neural Networks):

* Unscaled features lead to uneven gradients, causing slow or unstable convergence.

* Scaling normalizes the feature ranges, ensuring proportional updates for all weights.

* This results in faster convergence, stable training, and avoids oscillations or overshooting in gradient descent.

-Impact of Feature Scaling on Distance-Based Algorithms and Gradient Descent

1. Distance-Based Algorithms (e.g., KNN, SVM)

* Distance-based algorithms calculate similarity or distance between data points to make predictions. Examples include K-Nearest Neighbors (KNN) and Support Vector Machines (SVM).

* In KNN, the algorithm calculates the Euclidean distance or other distance metrics between points. If one feature has a large scale (e.g., salary in thousands) and another has a small scale (e.g., age 0–100), the large-scale feature will dominate the distance calculation, making the smaller-scaled features almost irrelevant.

* In SVM, feature scaling affects the margin calculation, as SVM tries to find a hyperplane that separates classes. Without scaling, features with larger ranges can bias the hyperplane, resulting in poor classification performance.

-Effect of Scaling:

* Scaling ensures all features contribute equally to the distance computation.

* Min-Max scaling or Standardization prevents any single feature from dominating the decision-making process.

* This leads to more accurate neighbor selection in KNN and better hyperplane separation in SVM.

2. Gradient Descent-Based Algorithms (e.g., Linear Regression, Logistic Regression, Neural Networks)

* Gradient descent is used to minimize a cost function by updating model weights iteratively. The step size depends on the gradient, which is influenced by the scale of the features.

* If features are on very different scales, the gradient in one direction can be much larger than in another. This causes the algorithm to oscillate along steep directions and converge slowly along shallow directions.

* For example, if one feature ranges from 0 to 1 and another from 0 to 1000, gradient updates for the large-scale feature will be huge, while the small-scale feature barely updates.

-Effect of Scaling:

* Scaling normalizes feature ranges, ensuring the gradient updates are proportional across all features.

* This leads to faster convergence, stable learning, and prevents overshooting during training.

* It is especially important for neural networks and algorithms using regularization, as feature magnitudes affect penalty terms.



Question 6:

 Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

Hint: Consider categorical variables with ordinal vs. nominal relationships.

Ans:-

* Definition:

-Categorical variables in a dataset contain discrete values or labels, such as colors, cities, or sizes. Most Machine Learning algorithms require numerical input, so categorical variables must be converted into numbers. Two common encoding techniques are Label Encoding and One-Hot Encoding.

-Label Encoding assigns a unique integer value to each category. For example, a “Size” feature with categories Small, Medium, Large can be encoded as Small → 0, Medium → 1, Large → 2.

-One-Hot Encoding converts each category into a binary vector, creating separate columns for each category. For example, a “Color” feature with categories Red, Green, Blue becomes Red → [1,0,0], Green → [0,1,0], Blue → [0,0,1].

* Importance of Encoding Categorical Variables:

-Encoding is important because Machine Learning models cannot process text directly. Proper encoding ensures that the model can learn patterns in categorical features accurately. Using the wrong encoding may introduce false relationships or bias, reducing the model’s performance. Encoding also improves compatibility with distance-based algorithms, gradient-based algorithms, and regularization techniques.

* Scope of Encoding:

-Encoding is used in a wide variety of Machine Learning tasks:

-Classification and regression problems where categorical input features exist.

-Distance-based algorithms like KNN and SVM, where numeric values affect distance calculations.

-Gradient descent-based algorithms such as Linear Regression, Logistic Regression, and Neural Networks, which require numeric inputs.

-Tree-based models like Decision Trees and Random Forests, where integer encoding is often sufficient.

-Encoding ensures accurate learning, fair feature representation, and efficient model training.

| **Aspect**                      | **Label Encoding**                                                      | **One-Hot Encoding**                                                                             |
| ------------------------------- | ----------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------ |
| **Definition**                  | Converts each category into a unique integer value.                     | Converts each category into a binary vector (0 or 1) across multiple columns.                    |
| **Output**                      | Single column with integer values.                                      | Multiple columns (one per category) with binary values.                                          |
| **Suitable For**                | Ordinal variables (where order matters).                                | Nominal variables (no natural order).                                                            |
| **Preserves Order**             | Yes, maintains hierarchy of categories.                                 | No, treats all categories equally without order.                                                 |
| **Memory Usage**                | Low, as only one column is used.                                        | High, increases dimensionality with more categories.                                             |
| **Computational Complexity**    | Low, faster to process.                                                 | Higher, requires more memory and computation.                                                    |
| **Risk of Misleading Model**    | High for nominal variables, as algorithm may assume false order.        | Low, as no ordinal assumption is introduced.                                                     |
| **Algorithm Compatibility**     | Works well with tree-based algorithms (Decision Trees, Random Forests). | Works well with distance-based and linear models (KNN, SVM, Linear Regression, Neural Networks). |
| **Interpretability**            | Easy to interpret single numeric value per category.                    | Slightly harder to interpret due to multiple binary columns.                                     |
| **Impact on Model Performance** | Good for ordinal data; may bias model for nominal data.                 | Good for nominal data; ensures fair representation of all categories.                            |

Practical Examples:

If a feature represents size, Label Encoding is preferred because the order matters.

If a feature represents color, One-Hot Encoding is preferred because categories are unrelated, and numeric encoding could mislead the model.

For algorithms like Decision Trees, Label Encoding may suffice even for nominal variables.

For KNN, SVM, linear regression, and neural networks, One-Hot Encoding is safer for nominal features to avoid introducing artificial ordinal relationships.

Question 7:

 Google Play Store Dataset

a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons? Dataset: https://github.com/MasteriNeuron/datasets.git (Include your Python code and output in the code box below.)

Ans:-

* To analyze the relationship between app categories and ratings in the Google Play Store dataset, we can perform the following steps:

* Load the dataset: Import the necessary libraries and load the dataset into a pandas DataFrame.

* Data Cleaning: Handle missing values and ensure data types are appropriate.

* Group by Category: Calculate the average rating for each app category.

* Visualization: Create a bar plot to visualize the average ratings by category.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv'
df = pd.read_csv(url)

# Data Cleaning
df = df.dropna(subset=['Category', 'Rating'])
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Group by Category and calculate the average rating
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
df_avg = category_ratings.reset_index()  # Convert to DataFrame

# Add a 'hue' column to allow using palette safely
df_avg['hue'] = df_avg['Category']  # temporary column for color mapping

# Colorful Bar Plot using viridis palette
plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Rating', hue='hue', data=df_avg,
            dodge=False, palette='viridis', legend=False)
plt.xticks(rotation=90)
plt.title('Average App Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

Explanation of the Code:

* Data Loading: The dataset is loaded from the provided URL using pandas' read_csv function.

* Data Cleaning: We drop rows with missing values in the 'Category' or 'Rating' columns and convert the 'Rating' column to numeric, coercing errors to NaN, which are then dropped.

* Grouping and Aggregation: The dataset is grouped by 'Category', and the mean of the 'Rating' is calculated for each group.

* Visualization: A bar plot is created using seaborn to visualize the average ratings for each category.

Possible Reasons for High or Low Average Ratings:

* High Ratings: Categories such as 'Education' and 'Health & Fitness' may have higher average ratings due to their focus on user well-being and educational value, which are often positively received by users.

* Low Ratings: Categories like 'Tools' or 'Productivity' might have lower average ratings if users have higher expectations for functionality and performance, leading to more critical reviews when these apps don't meet expectations.

Question 9:

Flight Price Prediction Dataset

a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)

Ans:-

a)Variation of Flight Prices with Days Left Until Departure:

* Flight prices generally increase as the departure date approaches, but the increase is not linear.

* When plotting flight prices against days left until departure, we observe:

* Prices are relatively low when booked well in advance.

* Around 1–2 weeks before the flight, prices surge exponentially, sometimes jumping significantly within a few days.

* This exponential increase is due to airlines raising fares as seats get limited and last-minute travelers have higher urgency.

-Best Booking Window:

* The ideal time to book flights is usually 2–3 weeks before departure, when prices are lower and stable.

* Booking too early might not always guarantee the lowest fares, and booking too late risks paying the exponentially higher last-minute prices.


```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

# Clean column names
df.columns = df.columns.str.strip()

# Detect columns automatically
days_col = df.columns[df.columns.str.contains('Days', case=False)][0]
price_col = df.columns[df.columns.str.contains('Price', case=False)][0]
source_col = df.columns[df.columns.str.contains('Source', case=False)][0]
dest_col = df.columns[df.columns.str.contains('Destination', case=False)][0]
airline_col = df.columns[df.columns.str.contains('Airline', case=False)][0]

# Part (a) - Price vs Days Left (Scatter plot)
plt.figure(figsize=(8,5))
sns.scatterplot(x=days_col, y=price_col, data=df, alpha=0.5, color='dodgerblue')
plt.title("Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left Until Departure")
plt.ylabel("Price (INR)")

# Annotate exponential surge
plt.axvline(x=14, color='red', linestyle='--', label='Price Surge Starts (~14 days)')
plt.text(14.5, df[price_col].max()*0.9, "Price Surge", color='red')
plt.legend()
plt.show()

# Average price by days left (Line plot)
avg_price_days = df.groupby(days_col)[price_col].mean().reset_index()
plt.figure(figsize=(8,5))
sns.lineplot(x=days_col, y=price_col, data=avg_price_days, marker='o', color='orange')
plt.title("Average Flight Price vs Days Left")
plt.xlabel("Days Left Until Departure")
plt.ylabel("Average Price (INR)")

# Highlight best booking window
best_window_start = avg_price_days[days_col].iloc[0]
best_window_end = 21  # around 2-3 weeks
plt.axvspan(best_window_start, best_window_end, color='green', alpha=0.2, label='Best Booking Window')
plt.legend()
plt.show()

# Part (b) - Airline Price Comparison for Delhi-Mumbai
route_df = df[(df[source_col]=='Delhi') & (df[dest_col]=='Mumbai')]
plt.figure(figsize=(10,5))
sns.boxplot(x=airline_col, y=price_col, data=route_df, color='lightblue', fliersize=5)
plt.title("Flight Price Comparison Across Airlines (Delhi-Mumbai)")
plt.ylabel("Price (INR)")
plt.xlabel("Airline")
plt.xticks(rotation=45)
plt.show()

# Average price per airline
avg_price_airline = route_df.groupby(airline_col)[price_col].mean().sort_values()
print("\nAverage Flight Price by Airline (Delhi-Mumbai):")
print(avg_price_airline)
```

b)Comparison of Prices Across Airlines (Delhi-Mumbai Route):

* Flight prices vary significantly depending on the airline, even for the same route.

-By analyzing the dataset:

* Some airlines are consistently cheaper (e.g., Indigo, SpiceJet)

* Other airlines are premium carriers with higher fares (e.g., Air India, Vistara)

-Reasons for Price Differences:

* Service Level and Amenities:

-Premium airlines offer more comfortable seating, better in-flight service, and additional facilities, which increases the ticket price.

-Budget airlines reduce costs by limiting amenities, resulting in lower fares.

* Brand Reputation and Reliability:

-Airlines with higher reliability, punctuality, and brand reputation often charge more.

* Flight Timing and Seat Availability:

-Some airlines may price differently depending on peak hours or limited seat availability, contributing to higher fares.

-Correlation Heatmap: Shows numeric variables like Job Satisfaction, Monthly Income, Years at Company, and their correlation with attrition.

-Boxplots: Compare distributions of Job Satisfaction and Monthly Income between employees who stayed and those who left.

-Countplots: Show proportion of employees leaving based on Overtime.


```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset directly from GitHub
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
df = pd.read_csv(url)

# Strip spaces from column names
df.columns = df.columns.str.strip()

# Check column names to identify attrition column
print("Columns in dataset:", df.columns)

# Specify the column representing employee leaving/attrition
# Replace 'left' with the actual column name if different
target_col = 'left'

# Convert target column to numeric
if df[target_col].dtype == 'object':
    df['Attrition_numeric'] = df[target_col].apply(lambda x: 1 if str(x).lower() in ['yes','y','1','left'] else 0)
else:
    df['Attrition_numeric'] = df[target_col]

# Part (a) - Key factors correlating with attrition
numeric_cols = df.select_dtypes(include='number').columns.tolist()
corr = df[numeric_cols].corr()['Attrition_numeric'].sort_values(ascending=False)
print("\nCorrelation with Attrition:")
print(corr)

# Visualize top factors using heatmap
key_features = ['JobSatisfaction','OverTime','MonthlyIncome','WorkLifeBalance','YearsAtCompany']
key_features = [col for col in key_features if col in df.columns]

plt.figure(figsize=(10,5))
sns.heatmap(df[key_features + ['Attrition_numeric']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Key Factors with Attrition")
plt.show()

# Boxplots / countplots for key drivers
if 'MonthlyIncome' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=target_col, y='MonthlyIncome', data=df)
    plt.title("Monthly Income vs Attrition")
    plt.show()

if 'OverTime' in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(x='OverTime', hue=target_col, data=df)
    plt.title("Overtime vs Attrition")
    plt.show()

if 'JobSatisfaction' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=target_col, y='JobSatisfaction', data=df)
    plt.title("Job Satisfaction vs Attrition")
    plt.show()

# Part (b) - Impact of Number of Projects
if 'NumProjects' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=target_col, y='NumProjects', data=df)
    plt.title("Number of Projects vs Attrition")
    plt.show()

    avg_projects = df.groupby(target_col)['NumProjects'].mean()
    print("\nAverage Number of Projects by Attrition:")
    print(avg_projects)
```

b)Impact of Number of Projects on Employee Attrition

* Visual Analysis:

-Boxplots comparing NumProjects for employees who stayed vs. left show that employees who left tend to have slightly more projects.

-This suggests that a higher workload may contribute to attrition.

*Average Comparison:

-Calculating the average number of projects per group:

-Employees who left → higher average number of projects
-Employees who stayed → lower average number of projects


-While the difference is noticeable, it is smaller than other factors like Overtime or Job Satisfaction.
